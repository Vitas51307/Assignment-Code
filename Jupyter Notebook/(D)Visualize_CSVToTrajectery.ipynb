## Read CharactorPositionTransverse.csv And Visualize
 * 儲存的 CharactorPosition.csv 需手動將其轉置並存成 CharactorPositionTransverse.csv 才可執行

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from numpy import genfromtxt

pos_data=[None]
pos_data = genfromtxt('D:/Vitas/台大/Port/CharactorPositionTransverse.csv', delimiter=',') #存在同資料夾下
x=pos_data[0]
z=pos_data[1]
y=pos_data[2]

ax = plt.axes(projection="3d")

ax.set_xlim3d(-7, 7)
ax.set_ylim3d(-7, 7)
ax.set_zlim3d(0, 7)

ax.set_xlabel('x')
ax.set_ylabel('z')
ax.set_zlabel('y')

ax.plot(x,y,z,c='r',marker='')
plt.show()
```
